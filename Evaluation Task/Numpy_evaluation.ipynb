```python
# You have to reduce the size of this dataset by 10 times.

# There are currently almost 5000 to 6000 thousand images (total: 60000) for each label in training and almost 1000 images (total: 10000) for each label in testing dataset.

# You are required to reduce this training dataset so that it will have exactly 600 images for each label (total: 6000) in training dataset and exactly 100 images for each label in testing dataset (total: 1000)

# new dataset should be stored into new_dataset variable name and should have exactly same format as the original dataset
```

```python
#your code here
import random

# Function to randomly select 'n' images for each label from the original dataset
def reduce_dataset(images, labels, n):
    reduced_images = []
    reduced_labels = []

    for label in set(labels):
        label_indices = [i for i, l in enumerate(labels) if l == label]
        selected_indices = random.sample(label_indices, n)

        for index in selected_indices:
            reduced_images.append(images[index])
            reduced_labels.append(label)

    return np.array(reduced_images), np.array(reduced_labels)

# Original dataset
dataset = keras.datasets.mnist.load_data()

# Reducing training dataset to have exactly 600 images for each label
reduced_training_images, reduced_training_labels = reduce_dataset(dataset[0][0], dataset[0][1], 600)

# Reducing testing dataset to have exactly 100 images for each label
reduced_testing_images, reduced_testing_labels = reduce_dataset(dataset[1][0], dataset[1][1], 100)

# Creating the new reduced dataset in the same format as the original dataset
new_dataset = ((reduced_training_images, reduced_training_labels), (reduced_testing_images, reduced_testing_labels))

# Validation of Task

# Checking the size of the new reduced dataset
print("There are", len(new_dataset[0][1]), "labels in training dataset for", len(new_dataset[0][0]), "images in training dataset")
print("There are", len(new_dataset[1][1]), "labels in testing dataset for", len(new_dataset[1][0]), "images in testing dataset")

# Checking the number of images per class in the new reduced dataset
print("There are", len(set(new_dataset[0][1])), "unique classes in training dataset")
print("Breakdown of each label in training dataset:")
print(collections.Counter(new_dataset[0][1]))

print("There are", len(set(new_dataset[1][1])), "unique classes in testing dataset")
print("Breakdown of each label in testing dataset:")
print(collections.Counter(new_dataset[1][1]))

# Checking the correctness of indexing of images and their labels
print("Randomly printing 10 images with labels from training dataset")
display_random_images_with_labels(new_dataset[0])

print("Randomly printing 10 images with labels from testing dataset")
display_random_images_with_labels(new_dataset[1])
```
